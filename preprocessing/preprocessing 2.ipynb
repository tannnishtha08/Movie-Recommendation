```python
meta['release_date'] = pd.to_datetime(meta['release_date'], errors='coerce')
```

```python
meta['year'] = meta['release_date'].dt.year
meta['year'].value_counts().sort_index()
```

```python
# we already have the data till 2016 and we need data after 2016 but from this data set we could only get the data from 2017 year
new_meta = meta.loc[meta.year==2017, ['genres','id', 'title','year']]
new_meta
```

```python
new_meta.info()
```

```python
new_meta['id'] = new_meta['id'].astype('int')
```

```python
data = pd.merge(new_meta, credits, on='id')
```

```python
pd.set_option('display.max_colwidth', 75)
data.head()
```

```python
data.genres.dtypes #data type is object but inside it contains list of dictionaries
```

```python
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x)) #convert str into the list
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
data.genres.dtypes
```

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return st.join(gen)

    
```

```python
data['genres_list'] = data['genres'].apply(lambda x : make_genresList(x))
data['genres_list']
```

```python
data['genres'][1]
```

```python
def get_actor1(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return (casts[0])

data['actor_1_name'] = data['cast'].apply(lambda x : get_actor1(x))

```

```python
data['actor_1_name'][0]
```

```python
def get_actor2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])

data['actor_2_name'] = data['cast'].apply(lambda x : get_actor2(x))
```

```python
data['actor_2_name'][0]
```

```python
def get_actor3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])

data['actor_3_name'] = data['cast'].apply(lambda x : get_actor3(x))
```

```python
data['cast'][0]
```

```python
def get_director(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return st.join(dt)

data['director_name'] = data['crew'].apply(lambda x : get_director(x))
```

```python
data.columns
```

```python
movie = data.loc[:, ['director_name','actor_1_name', 'actor_2_name', 'actor_3_name','genres_list','title']]
```
